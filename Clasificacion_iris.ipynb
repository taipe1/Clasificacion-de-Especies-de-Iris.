```python
# NumPy para conservar datos
import numpy as np
# Matplotlib es para mostrar nuestros datos
%matplotlib inline
import matplotlib.pyplot as plt
# Pandas es nuestra base de datos
import pandas as pd
# Sklearn para crear nuestro modelo
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
# Mglearn es de Introducción al ML con Python
import mglearn
```



### Leer los datos


```python
iris_dataset = load_iris()
print("cargamos la clave iris_dataset:\n", iris_dataset.keys())
```

```python
print("Clases de Iris:\n", iris_dataset["target_names"])
```

### Dado que nuestro proceso implica entrenamiento y pruebas, debemos dividir nuestro conjunto de datos, que se puede ejecutar mediante el siguiente código


```python
X_train, X_test, Y_train, Y_test = train_test_split(iris_dataset["data"], iris_dataset["target"], random_state=0)
print("X_train shape:", X_train.shape)
print("Y_train shape:", Y_train.shape)
print("X_test shape:", X_test.shape)
print("Y_test shape:", Y_test.shape)
```

### Veamos los datos


```python
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
pd.plotting.scatter_matrix(iris_dataframe, 
                           c=Y_train, 
                           figsize=(10,10), 
                           marker='x', 
                           hist_kwds={"bins":20}, 
                           s=60, 
                           alpha=0.8, 
                           cmap=mglearn.cm3)
```

### Building k- Vecinos más cercanos

```python
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, Y_train)
print("Puntaje de datos de prueba: {:.4f}".format(knn_model.score(X_test, Y_test)))
```
